```python
df['age'] = df['age'].round().astype('Int64')
```

```python
df.dtypes
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
import joblib

cat_col=["gender","diabetic","smoker","region"]

encoders = {}

for col in cat_col:
    lab = LabelEncoder()
    x[col] = lab.fit_transform(x[col])
    encoders[col] = lab   # save encoder for later use
    joblib.dump(lab,f"label_encoder_{col}.pkl")
```

```python
x
```

```python
encoders
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(
    n_estimators=300,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    n_jobs=-1
)

# model=RandomForestRegressor(n_estimators=100,n_jobs=-1,verbose=0)#n_jobs=-1 → Random Forest will use all CPU cores to build trees in parallel.For huge datasets, it can save a lot of time.
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
data=({"Actual_value":y_test,"Predicted_value":y_pred,"Difference":y_test-y_pred})
df1=pd.DataFrame(data)
df1
```

```python
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,y_pred)
mae
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape=mean_absolute_percentage_error(y_test,y_pred)
mape
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
from sklearn.metrics import root_mean_squared_error
rmse=root_mean_squared_error(y_test,y_pred)
rmse
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```

```python
train_score = model.score(x_train, y_train)
test_score  = model.score(x_test, y_test)

print(train_score)
print(test_score)
```

```python
train_rmse=root_mean_squared_error(y_train,model.predict(x_train))
test_rmse=root_mean_squared_error(y_test,model.predict(x_test))

print("Train RMSE:", train_rmse)
print("Test RMSE :", test_rmse)
```
